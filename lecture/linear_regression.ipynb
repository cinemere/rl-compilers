# Linear regression

## Task: predict temperature based on the number of chirps per minute

#### ***Reference:*** Dolbear's law

In 1890, Amos Dolbear determined that there was a fixed relationship between the number of chirps per minute of the snowy tree cricket and the ambient temperature.





```python
def dolbears_law(N):
    # return 50 + (N - 40)/4 F
    return 10 + 5 * (N - 40) / 36
```

## Create the data:



```python
import numpy as np

# T = np.random.uniform(5, 40, 50)
T = np.arange(5, 30)
```

```python
# N = 4 * T - 360 + np.random.randn(len(T)) * 5
N = ((36 * T - 160)/5 + np.random.randn(len(T)) * 5).astype(int)
```

```python
import matplotlib.pyplot as plt 

plt.scatter(T, N)
plt.title("Collected data")
plt.xlabel('temperature')
plt.ylabel('chirps per minute')
```

```python
import pandas as pd

df = pd.DataFrame({
    "temperature": T,
    "chrips_per_minute": N
})
```
